## 1. Obtención e inspección de archivos

Los datos de encuestas fueron obtenidos como parte de un trabajo de campo, y por limitaciones tecnológicas, solo se cuenta con la información en archivos de texto plano. Cada archivo corresponde a una de las variables antes mencionadas y se puede garantizar que todos relacionan a las personas encuestadas en el mismo orden (i.e., la primera línea de todos los archivos corresponde a datos de la primera persona encuestada, la segunda línea a la segunda persona, etc.).

Los archivos con los que se cuenta son:
* *edades.txt*: un archivo ordenado con las edades de cada persona de la población encuestada 
* *género.txt*: un archivo ordenado con el género de cada persona de la población encuestada 
* *estado_civil.txt*: un archivo ordenado con el estado civil de cada persona de la población encuestada 
* *escolaridad.txt*: un archivo ordenado con el nivel de escolaridad de cada persona de la población encuestada 
* *estrato.txt*: un archivo ordenado con el estrato socio-económico de cada persona de la población encuestada
* *region.txt*: un archivo ordenado con la región de proveniencia de cada persona de la población encuestada

**Misión 1**

Poner en variables de Python la información que está en los archivos listados arriba.

* *Versión base*: escribe línea a línea el código necesario para poner en variables de Python la información de cada uno de los archivos.

* *Milla extra*: escribe una función que sea capaz de hacer lo anterior para un archivo y úsala en un ciclo para automatizar la lectura de los diferentes archivos.

```python
def archivo_a_lista(path):
    f = open(path, mode="r", encoding="utf-8") 
    lines = f.read().splitlines()
    f.close()
    return lines

def trasponer_dicc(diccionario):
    newdict = {}
    primer_llave = next(iter(diccionario))
    for v in range(len(diccionario[primer_llave])):
        atributos = []
        for k in diccionario:
            atributos.append(diccionario[k][v])
        newdict[v] = atributos
    return newdict

def convert_int(lista):
    newlist = []
    for i in lista:
        newlist.append(int(i))
    return newlist

def convert_float(lista):
    newlist = []
    for i in lista:
        newlist.append(float(i))
    return newlist

#%%
cadena_archivos = ["./Archivos/edad.txt","./Archivos/genero.txt",\
                   "./Archivos/estado_civil.txt","./Archivos/escolaridad.txt",\
                   "./Archivos/estrato.txt","./Archivos/region.txt",\
                   "./Archivos/promedio.txt"]
cadena_tipodata = ["Edad", "Genero", "Estado_civil", "Escolaridad", "Estrato",\
                   "Region", "Promedio"]

data = {}

for i in range(len(cadena_archivos)):
    data[cadena_tipodata[i]] = archivo_a_lista(cadena_archivos[i])

data[cadena_tipodata[0]] = convert_int(data[cadena_tipodata[0]])
data[cadena_tipodata[6]] = convert_float(data[cadena_tipodata[6]])
  
data_traspuesta = trasponer_dicc(data)
data_traspuesta
```

**Misión 2**

Reportar las principales características en los datos.
* *Versión base*: utilizar métricas básicas como conteos o promedios por categorías en los datos.
* *Milla extra*: utilizar visualizaciones simples como diagramas de barras, de pastel u otros.

```python
def filtrar_data_dicc (lista_filtros,data):
    resultado_filtro = []
    
    for filtro in lista_filtros:
       for linea in data:
            if linea == filtro:
                resultado_filtro.append(data[linea])
                break
    return resultado_filtro

def medidas_descriptivas(param_list):
    import numpy as np
    lista_descriptivas = ["Tamaño","Moda","Media","Mediana","Varianza","Desviacion"]
    dicc_medidas = {}
    
    for i in lista_descriptivas:
        if i == "Tamaño":
            dicc_medidas[i] = len(param_list)
        elif i == "Moda":
            dicc_medidas[i] = moda(param_list)
        elif i == "Media":
            dicc_medidas[i] = round(np.mean(param_list),4)
        elif i == "Mediana":
            dicc_medidas[i] = np.median(param_list)
        elif i == "Varianza":
            dicc_medidas[i] = round(np.var(param_list),4)
        else:
            dicc_medidas[i] = round(np.std(param_list),4)
    return dicc_medidas

def moda(param_list):
    valor_moda = 0
    for i in param_list:
        if counter_val(param_list,i)> valor_moda:
            valor_moda = i
    return valor_moda

def counter_val(param_list, val):
    count = 0
    for i in param_list:
        if i == val:
            count += 1
    return count

#%%
filtro_continuas=["Edad","Promedio"]
data_filtrada = filtrar_data_dicc(filtro_continuas,data)

medidas_edad = medidas_descriptivas(data_filtrada[0])
medidas_promedio = medidas_descriptivas(data_filtrada[1])

print("Medidas Edad", medidas_edad)
print("Medidas Promedio", medidas_promedio)
    

import matplotlib.pyplot as plt

# Inicializar layout Histogramas
fig, axs = plt.subplots(7, 1, figsize=(6,15), sharey=False, tight_layout=True)
colors = ["#741b47", "#a64d79", "#c27ba0", "#d5a6bd", "#d5a6bd", "#c27ba0", "#a64d79"]

for i in range(len(cadena_tipodata)):
    axs[i].hist(data[cadena_tipodata[i]], bins=10, color=colors[i])
    axs[i].title.set_text("Histograma " + cadena_tipodata[i])
plt.show()
    
fig, axs1 = plt.subplots(1, 2, figsize=(5,3), sharey=False, tight_layout=True)

# Inicializar layout Boxplots
for i in range(len(data_filtrada)):
    axs1[i].boxplot(data_filtrada[i])
    axs1[i].title.set_text(filtro_continuas[i])
plt.show()
```

**Misión 4**

Desarrollar un algoritmo que permita conocer la población que sería becada según los criterios propuestos por el *primer grupo* mencionado arriba.

```python
def agregar_id(atributos):
    newlist = []
    individuos = len(atributos[0])
    for v in range(individuos):
        elementos = []
        elementos.append(v)
        for k in atributos:
            elementos.append(k[v])
        newlist.append(elementos)
    return newlist

def param_asignacion(datos, indice_param):
    list_param = []
    for i in datos:
        list_param.append(i[indice_param])
    return list_param

def unique(param_list):
    unique_list = []
    for x in param_list:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list

def asignacion_porcentaje(param_list,ind_sort,porcentaje,becas_disponibles):
    tamano_grupo = len(param_list)
    cant_asig = round(porcentaje * tamano_grupo)
    param_list.sort(key=lambda x: x[ind_sort])
    param_list.reverse()
    
    if becas_disponibles < cant_asig:
        cant_asig = becas_disponibles
    
    asignados =[]
    assig = 0
    
    while assig < cant_asig:
        asignados.append(param_list[assig])
        assig += 1
    return asignados

def excluir_data (lista_filtros,data):
    newData = data.copy()
    for d in lista_filtros:
        if d in data:
            newData.remove(d)
    return newData

def extraer_listado(param_list, indice):
    newList = []
    for ind in param_list:
        newList.append(ind[indice])
    return newList

#%%
    
filtro_metodo1 = ["Estrato","Promedio"]
data_metodo1 = filtrar_data_dicc(filtro_metodo1,data)
data_metodo1 = agregar_id(data_metodo1)

def regla1(data_regla1, indice_regla, num_becas):
    elementos_parametro = sorted(unique(param_asignacion(data_regla1,indice_regla)))
    becas_asignadas = []
    becas_disponibles = num_becas
    
    while becas_disponibles > 0:
        salir = 0
        for j in elementos_parametro:
            subgrupo = []
            for k in data_regla1:
                if j == k[indice_regla]:
                    subgrupo.append(k)
                    
            if round(len(subgrupo)*0.02) < 1:
                salir += 1

            asignados_subgrupo = asignacion_porcentaje(subgrupo,2,0.02,becas_disponibles)
            becas_disponibles = becas_disponibles - len(asignados_subgrupo)
            becas_asignadas.extend(asignados_subgrupo)
            data_regla1 = excluir_data(asignados_subgrupo,data_regla1)
        
        if salir == len(elementos_parametro):
            break
                
    return extraer_listado(becas_asignadas,0)

#Ejemplo
resultado1 = sorted(regla1(data_metodo1,1,20))
resultado1
```

**Misión 5**

Desarrollar un algoritmo que permita conocer la población que sería becada según los criterios propuestos por el *segundo grupo* mencionado arriba.

```python
def combinar_param(elementos1, elementos2):
    comb = []
    for i in elementos1:
        for j in elementos2:
            comb.append([i,j])
    return comb

def asignar_estatico(param_list,ind_sort,becas_asignar):
    param_list.sort(key=lambda x: x[ind_sort])
    param_list.reverse()
      
    asignados =[]
    assig = 0
    
    if len(param_list) < becas_asignar:
        becas_asignar = len(param_list)
        
    while assig < becas_asignar:
        asignados.append(param_list[assig])
        assig += 1
    return asignados

#%%
filtro_metodo2 = ["Genero","Region","Promedio"]
data_metodo2 = filtrar_data_dicc(filtro_metodo2,data)
data_metodo2 = agregar_id(data_metodo2)

def regla2(data_regla2,num_becas):
    import math as mt
    elementos1_parametro = sorted(unique(param_asignacion(data_regla2,1)))
    elementos2_parametro = sorted(unique(param_asignacion(data_regla2,2)))
    becas_asignadas = []
    
    combinaciones = combinar_param(elementos1_parametro,elementos2_parametro)
    becas_por_subgrupo = mt.floor(num_becas/len(combinaciones))
    
    for j in combinaciones:
        subgrupo = []
        for k in data_regla2:
            if j[0] == k[1] and j[1] == k[2]:
                subgrupo.append(k)
        asignados_subgrupo = asignar_estatico(subgrupo,3,becas_por_subgrupo)
        becas_asignadas.extend(asignados_subgrupo)

    return extraer_listado(becas_asignadas,0)

#Ejemplo
resultado2 = sorted(regla2(data_metodo2,20))
resultado2
```

```python
def filtrar_rango(param_list,ind_rango,liminf_rango, limsup_rango):
    newList = []
    
    for item in param_list:
        if item[ind_rango] in range(liminf_rango, limsup_rango+1):
            newList.append(item)
    return newList    

#%%
def regla3(rango_edad_liminf, rango_edad_limsup, criterio, becas, proporcion_subgrupo):
    import math as mt
    filtro_metodo3 = ["Edad",criterio,"Promedio"]
    data_metodo3 = filtrar_data_dicc(filtro_metodo3,data)
    data_metodo3 = agregar_id(data_metodo3)
    
    data_metodo3 = filtrar_rango(data_metodo3,1,rango_edad_liminf, rango_edad_limsup)
    
    elementos_parametro = sorted(unique(param_asignacion(data_metodo3,2)))
    becas_asignadas = []
        
    try:
        while len(proporcion_subgrupo) != len(elementos_parametro):
            return print("Debes proporcionar todos los % becas para cada sub-categoria del criterio "\
                         + criterio)
        
        validar_proporciones = sum(item for item in proporcion_subgrupo.values())
        
        while validar_proporciones != 100:
            return print("Las proporciones para el criterio deben sumar 1")
        
        for j in elementos_parametro:
            if j not in proporcion_subgrupo:
                return print("No incluiste la sub-categoria " +j)
        
    except TypeError:
        print("Cerciorese que las proporciones sean un diccionario donde las llaves son las sub-categoria del criterio escogido")
    
    for j in elementos_parametro:
        subgrupo = []
        for k in data_metodo3:
            if j == k[2]:
                subgrupo.append(k)
                    
        becas_por_subgrupo = mt.floor(becas*proporcion_subgrupo[j]/100)
        asignados_subgrupo = asignar_estatico(subgrupo,3,becas_por_subgrupo)
        becas_asignadas.extend(asignados_subgrupo)
    
    return extraer_listado(becas_asignadas,0)

#Ejemplo
proporciones_genero = {'femenino':40, 'masculino':20, 'no binario':20, 'otro':20 }
proporciones_estrato = {'1':20, '2':20, '3':20, '4':20, '5':20}
proporciones_region = {'Andina':20, 'Pacifica':20, 'Caribe':20, 'Orinoquia':20, 'Amazonia':20}
resultado3 = sorted(regla3(20,30,"Region",20, proporciones_region))
resultado3   
    
```
